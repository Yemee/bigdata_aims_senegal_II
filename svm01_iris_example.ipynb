<h1> svm_iris_example</h1>

<p>An example of the nearest neighbor algorithm application using the well known iris dataset</p>
<p>First we import the dataset</p>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data[(iris.target==0)|(iris.target==1),0:3:2]
y = iris.target[(iris.target==0)|(iris.target==1)]
```

```python
plt.scatter(X.T[0],X.T[1],c=y)
```

```python
from sklearn.preprocessing import scale
X = scale(X)
```

```python
from sklearn.svm import LinearSVC

lsvc = LinearSVC(C=100.0)
lsvc.fit(X,y)
```

<p>Mapping the parameters space and visualizing the result.. and check the difference</p>

```python
xx,yy = np.meshgrid(np.linspace(0.0,10.0,201),np.linspace(0.0,10.0,201))
predict = lsvc.predict(np.c_[xx.ravel(),yy.ravel()])
Z_lsvc = predict.reshape(np.shape(xx))
```

```python
plt.figure(figsize=(14,7))

plt.subplot(121)
plt.contourf(xx,yy,Z_lsvc,alpha=0.4)
plt.scatter(X.T[0],X.T[1],c=iris['target'])
plt.axis('scaled')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('SVC',fontsize=20)
plt.xlim(3.0,9.0)
plt.ylim(0.0,8.0)
```
